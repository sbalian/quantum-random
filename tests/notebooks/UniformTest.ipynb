# Visualise and test if the quantum randoms are uniformly distributed in $\left[0.0, 1\right)$

Unlike the pytests, this notebook queries the real API.

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats

import qrandom
import random
```

```python
mpl.rcParams["figure.dpi"] = 120
```

```python
numbers = [qrandom.random() for _ in range(10000)]
```

```python
py_numbers = [random.random() for _ in range(10000)]
```

```python
plt.figure(figsize=(8, 3))
plt.subplot(1, 2, 1)
plt.hist(numbers, edgecolor="black")
plt.title("Quantum random numbers")
plt.subplot(1, 2, 2)
plt.hist(py_numbers, edgecolor="black")
plt.title("Python standard random numbers")
plt.tight_layout();
```

Run Kolmogorov-Smirnov

```python
stats.kstest(numbers, "uniform")
```

```python
stats.kstest(py_numbers, "uniform")
```
